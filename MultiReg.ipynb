```python
import pandas as pd
```

```python
data = pd.read_csv('CarPrice_Assignment.csv')
data.head()
```

```python
data.dtypes
```

```python
numeric_columns = data.select_dtypes(include=['int64', 'float64'])
```

```python
data["enginetype"].unique()
```

```python
data["carbody"].unique()
```

```python
data = pd.get_dummies(data, columns=['fueltype'])
```

```python
data.drop("fueltype_gas",axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
label_encoder = preprocessing.LabelEncoder()
data['carbody'] = label_encoder.fit_transform(data['carbody'])
data['enginetype'] = label_encoder.fit_transform(data['enginetype'])
```

```python
X = data[["horsepower","fueltype_diesel","enginesize","enginetype","carbody"]]
Y = data["price"]
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import math
```

```python
print('mean_squared_error: ',mean_squared_error(Y_test,y_pred))
print('mean_absolute_error: ',mean_absolute_error(Y_test,y_pred))
print('root_mean_squared_error: ',math.sqrt(mean_squared_error(Y_test,y_pred)))
math.sqrt(mean_squared_error(Y_test,y_pred))
```
